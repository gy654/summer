```python
import pandas as pd
import numpy as np
admissions = pd.read_csv('ADMISSIONS.csv.gz',  compression='gzip',error_bad_lines=False)
note = pd.read_csv('NOTEEVENTS.csv.gz',  compression='gzip',error_bad_lines=False)
patients = pd.read_csv('PATIENTS.csv.gz',  compression='gzip',error_bad_lines=False)
```

```python
note = note[note['CATEGORY']=='Discharge summary']
```

```python
print(len(admissions), admissions.columns)
print(len(patients),patients.columns)
print(len(note), note.columns)
```

```python

display(admissions)
display(patients)
display(note)

#admissions.ETHNICITY.value_counts().plot(kind = 'bar')

```

```python
from datetime import datetime

patients['DOD']= pd.to_datetime(patients['DOD']).dt.date
patients['DOB']= pd.to_datetime(patients['DOB']).dt.date
admissions['ADMITTIME']= pd.to_datetime(admissions['ADMITTIME']).dt.date
admissions['DISCHTIME']= pd.to_datetime(admissions['DISCHTIME']).dt.date
```

```python
#update
pat_sub_gen_age = patients[['SUBJECT_ID', 'GENDER']]
adm_id_hadm_adm = admissions[['SUBJECT_ID','HADM_ID','ADMITTIME', 'DISCHTIME','ETHNICITY']]
patient_admission = pd.merge(pat_sub_gen_age,adm_id_hadm_adm,on='SUBJECT_ID', how='left')

patient_admission['readmit_count'] = patient_admission.groupby('SUBJECT_ID')['SUBJECT_ID'].transform('count')
patient_admission = patient_admission.sort_values(['SUBJECT_ID','ADMITTIME'])
```

```python
patient_admission.reset_index(drop=True, inplace =True)
```

```python
patient_admission
```

```python
length = len(patient_admission)
readmission = []
interval = []
for i in range(length-1):
    if  patient_admission.iloc[i+1]['SUBJECT_ID']== patient_admission.iloc[i]['SUBJECT_ID']:
        diff = patient_admission.iloc[i+1]['ADMITTIME']-patient_admission.iloc[i]['DISCHTIME']
        interval.append(diff)
        if(diff < pd.Timedelta(31,'D')):
            readmission.append(1)
        else:
            readmission.append(0)
    else:
        readmission.append(0)
        interval.append(np.nan)
readmission.append(0)
interval.append(np.nan)

patient_admission['INTERVAL']= interval
patient_admission['readmission']= readmission
```

```python
positive = patient_admission[patient_admission['readmission']==1]
```

```python
positive
```

```python
# join the note
patient_admission_note = pd.merge(patient_admission, note[['HADM_ID', 'TEXT']], on=['HADM_ID'], how='inner').sort_values('SUBJECT_ID')

```

```python
patient_admission_note
```

```python
patient_admission_note = patient_admission_note.groupby('HADM_ID').agg({'SUBJECT_ID':'first', 
                                                     'GENDER':'first',
                                                     'ADMITTIME':'first',
                                                     'DISCHTIME':'first',
                                                     'ETHNICITY':'first',
                                                     'readmit_count':'first',
                                                     'readmission':'first',
                                                     'INTERVAL':'first',
                                                     'TEXT': ', '.join}).reset_index()
```

```python
patient_admission_note
```

```python
positive_notes = patient_admission_note[patient_admission_note['readmission']==1]
```

```python
positive_notes
```

```python
patient_admission_note.to_csv('processed.csv')
```

```python
print("Number of positive 30-day readmission samples:",len(df[df['readmission']==1])," out of total admission samples:",len(df))

print("Proportion of positive 30-day readmission samples in all ethnic groups:", 
      len(df[df['readmission']==1]) / len(df))
print("Proportion of positive 30-day readmission samples in WHITE ethnic group:", 
      len(df[df['ETHNICITY']=='WHITE'][df['readmission']==1]) / len(df[df['ETHNICITY']=='WHITE']))
print("Proportion of positive 30-day readmission samples in ASIAN ethnic groups:", 
      len(df[df['ETHNICITY'].str.contains('ASIAN')][df['readmission']==1]) / len(df[df['ETHNICITY'].str.contains('ASIAN')]))
print("Proportion of positive 30-day readmission samples in BLACK ethnic group:", 
      len(df[df['ETHNICITY']=='BLACK/AFRICAN AMERICAN'][df['readmission']==1]) / len(df[df['ETHNICITY']=='BLACK/AFRICAN AMERICAN']))
print("Proportion of positive 30-day readmission samples in Males:", 
      len(df[df['GENDER']=='M'][df['readmission']==1]) / len(df[df['GENDER']=='M']))
print("Proportion of positive 30-day readmission samples in Females:", 
      len(df[df['GENDER']=='F'][df['readmission']==1]) / len(df[df['GENDER']=='F']))

print(df.ETHNICITY.value_counts())
```
